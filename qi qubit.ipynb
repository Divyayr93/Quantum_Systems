```python
import cmath
import numpy as np
```

```python
zero_ket = np.array([1+0j, 0+0j])
```

```python
print(zero_ket)
```

```python
from qiskit.visualization import plot_bloch_multivector
```

```python
plot_bloch_multivector(zero_ket)
```

```python
psi_ket = np.array([1/cmath.sqrt(2)*1+0j, 1/cmath.sqrt(2)*1+0j])
```

```python
plot_bloch_multivector(psi_ket)
```

```python
psi_minus_ket = np.array([1/cmath.sqrt(2)*1+0j, -1/cmath.sqrt(2)*1+0j])
```

```python
plot_bloch_multivector(psi_minus_ket)
```

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.quantum_info import Statevector
```

```python
qr = QuantumRegister(3,'q')
cr = ClassicalRegister(3,'c')
qc = QuantumCircuit(qr, cr)
sv1 = Statevector(qc)
```

```python
from qiskit.visualization import array_to_latex
```

```python
from qiskit.visualization import plot_bloch_vector
```

```python
plot_bloch_multivector(sv1)
```

```python
array_to_latex(sv1)
```

```python
qc = QuantumCircuit(qr, cr)
qc.h(qr)
sv2=Statevector(qc)
qc.draw('mpl')
```

```python
array_to_latex(sv2)
```

```python
# 011
qc = QuantumCircuit(qr, cr)
qc.h(qr[0])
qc.x(qr[1])
qc.x(qr[2])
sv2=Statevector(qc)
qc.draw('mpl')
```

```python
array_to_latex(sv2)
```
